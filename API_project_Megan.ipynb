Question:1
    Use the API to get all available data for the GDP per capita, PPP (constant 2017 international $) indicator. 
    Hint: this indicator has code "NY.GDP.PCAP.PP.KD". 
    Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
endpoint = "http://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD"
```

```python
params = {"format": "json", "per_page": "20000"}
```

```python
res = requests.get(endpoint, params=params)
```

```python
res
```

```python
res = res.json()
```

```python
df = pd.json_normalize(res[1])
```

```python
cleaned_gdp = df[["date","value","country.value"]]
```

```python
cleaned_gdp.columns = ["year","gdp","country"]
```

```python
cleaned_gdp
```
